```python
mean, var, skew, kurt = binom.stats(10,0.5,moments='mvsk')
print(f'''
Mean: {mean}
Variance: {var}
Skew: {skew}
Kurtosis: {kurt}
''')
```

1. PDF: Probability density function  
2. PMF: Probability mass function  
3. CDF: Cumulative distribution function
4. PPF: Percent point function (inverse of cdf — percentiles).

```python
cdf = binom.cdf(15,30,0.5)
ppf = binom.ppf(0.5,30,0.5)
print(f"cdf is {cdf} ppf is {ppf}")
```
